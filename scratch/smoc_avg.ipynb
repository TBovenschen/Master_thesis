```python
import numpy as np
import xarray as xr
```

```python
path_dataSMOC = '/data/oceanparcels/input_data/CMEMS/GLOBAL_ANALYSIS_FORECAST_PHY_001_024_SMOC/'



mean_vel = xr.open_mfdataset(path_dataSMOC+'*.nc',combine='by_coords')
means = mean_vel.sel(latitude=slice(55,65), longitude=slice(-65,-45)).resample(time='1M', loffset='-15D').mean(dim='time')

means.interp(latitude=np.linspace(55,65,121), longitude = np.linspace(-65,-45, 241))
means.to_netcdf('SMOC_monthly_mean.nc')
```
